```python
from nsepy import get_history
from datetime import date, timedelta
import pandas as pd
```

```python
# Taking initial dataset as NIFTY 100. Run different analysis on this and then figure out top 20 stocks. 
# These 20 stocks are picked depending on different parameters like past 20 days trend and their current price. 
# It might be that I do not want any stock over the 500 Rs. price point as LCP.

nifty_100 = pd.read_csv("./Data/ind_nifty100list.csv")
```

```python
nifty_100
```

```python
current_date = date.today()
five_years = 365 * 5
five_years_ago_date = current_date - timedelta(days=five_years)
```

```python
# Stock market data for past 5 years so that we don't have to call nsepy again and again

stock_market_data = []

for index, row in nifty_100.iterrows():
    print(row["Symbol"])
    stock_data = get_history(symbol=row["Symbol"],
                  start=five_years_ago_date,
                  end=current_date)
    stock_market_data.append(stock_data)
```
